# Configurando Subplots de Análise de Vendas por Loja em 2022

Este notebook apresenta uma análise visual da variação no número de vendas mensais para quatro lojas (A, B, C e D) durante o ano de 2022.

---

## Objetivo

Visualizar o comportamento das vendas ao longo do ano para cada loja separadamente, facilitando a identificação de tendências e sazonalidades.

---

## Importação das bibliotecas necessárias


```python

import pandas as pd
import matplotlib.pyplot as plt
```


## Preparação dos Dados

Definimos as vendas mensais para cada loja em 2022 e construímos um DataFrame para manipulação.


```python

# Dados
lojas = ['A', 'B', 'C', 'D']
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)

# Transpondo para facilitar o plot
df_transposto = df.T
```


## Visualização Gráfica

Criamos gráficos separados para cada loja para facilitar a análise individual das vendas ao longo dos meses. Utilizamos cores distintas para cada loja, com personalização dos eixos, títulos e grades.



```python

# Cores para cada loja
cores = ['blue', 'green', 'orange', 'red']

# Criando subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 8))
fig.suptitle('Variação no Número de Vendas por Loja em 2022', fontsize=18, fontweight='bold')

# Plotando com personalização
for i, ax in enumerate(axs.flat):
    loja = lojas[i]
    ax.plot(df_transposto.index, df_transposto[loja], color=cores[i], linewidth=2.5)
    ax.set_title(f'Loja {loja}', fontsize=12, loc='left')
    ax.set_xlabel('Mês', fontsize=10)
    ax.set_ylabel('Vendas', fontsize=10)
    ax.tick_params(axis='x', rotation=45)
    ax.grid(True, linestyle='--', alpha=0.5)

# Layout organizado
plt.tight_layout(rect=[0, 0, 1, 0.95])  # espaço para o título

# Salvando a figura em PNG com fundo branco e 300 dpi
fig.savefig('vendas_lojas_2022.png', dpi=300, facecolor='white')

plt.show()
```
